```python
import pytablewriter
import io
import os

#df_data_3 = pd.read_csv("Oui_MANA_articles.tsv",delimiter='/',encoding='utf-8')
df_data_3 = pd.read_csv("Oui_MANA_articles",delimiter='/',encoding='utf-8')



writer = pytablewriter.MarkdownTableWriter()
writer.table_name = "Articles OUI MANA"
writer.header_list = list(df_data_3.columns.values)
indexes_to_delete=[11,10,9,8,7,5,4,1]
writer.header_list[0]="Number priority ranking article"
writer.header_list[3]="Number alerting entities detected"
for index in indexes_to_delete:
    del writer.header_list[index]
    
writer.value_matrix = df_data_3.values.tolist()


for pointer in writer.value_matrix:
    
    if type(pointer[5])!=float and type(pointer[4])!=float and type(pointer[6])!=float:
        
        article_highlighted=pointer[6]

        sentences_list=pointer[5].split(";")
        if '' in sentences_list:
            sentences_list.remove('')

        for sentence in sentences_list:
            article_highlighted=article_highlighted.replace(sentence,'<mark>'+sentence+'</mark>')  

        keywords_list=pointer[4].split(";")
        if '' in keywords_list:
            keywords_list.remove('')

        for keyword in keywords_list:
            article_highlighted=article_highlighted.replace(keyword,'<mark style="background-color: red">'+keyword+'</mark>')
        
        article_highlighted=article_highlighted.replace('$','dollars')

        pointer[6]="<details open> <summary>Full Article</summary>  <br> <details>"+article_highlighted
    else: 
        pointer[6]="<details open> <summary>Full Article</summary>  <br> <details>"+pointer[6]
    
    for index in indexes_to_delete:
        del pointer[index]

with open("articles_OUI_MANA_markdown","w+") as x:
    writer.stream = x
    writer.write_table() 
    
from IPython.display import display, Markdown, Latex
display(Markdown("articles_OUI_MANA_markdown"))
```
